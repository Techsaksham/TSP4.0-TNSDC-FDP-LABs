## Histograms and Density Plots

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data (randomly generated)
data = np.random.randn(1000)  # Generate 1000 random values from a normal distribution

# Create histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, density=False, alpha=0.75, color='blue')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Data')
plt.grid(True)
plt.show()
```

Explanation:

1. plt.hist(data, bins=30, density=False, alpha=0.75, color='blue')

   creates a histogram:
   
   1.1 data: The array of values to be plotted.

   1.2 bins=30: Number of bins (intervals) to divide the data range into.

   1.3 density=False: If True, the histogram will show a probability density rather than counts.

   1.4 alpha=0.75: Transparency of bars (0 for transparent, 1 for opaque).

   1.5 color='blue': Color of bars in the histogram.
3. plt.xlabel(), plt.ylabel(), plt.title(): Add labels and a title to the plot.

4. plt.grid(True): Optionally add grid lines to the plot for better readability.

## Density Plots Using Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data (using the Iris dataset for illustration)
data = sns.load_dataset("iris")

# Create density plot
plt.figure(figsize=(8, 6))
sns.kdeplot(data['sepal_length'], shade=True, color='g', label='Sepal Length')
sns.kdeplot(data['sepal_width'], shade=True, color='b', label='Sepal Width')
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Density Plot of Sepal Measurements')
plt.legend()
plt.show()
```
